# 1 - Preprocess
Take the combined Tweet and prouce two dataset those with excite and those with dismay

```python
import pandas as pd
results = pd.read_csv('combined_codes.csv')
```

```python
results['dismay'] = results['final_code'].str.contains('dismay').values.astype(int)
results['excite'] = results['final_code'].str.contains('excite').values.astype(int)
```

```python
results[results['tweet_text'].str.contains('RT')]
```

```python
excite_col = ['tweet_text','excite']
dismay_col = ['tweet_text','dismay']
```

```python

results.to_csv('excite.csv', columns = excite_col)
results.to_csv('dismay.csv', columns = dismay_col)
```
