# 2 -  text error analysis


```python
def filter_feature(data):
    feature_error = data
    feature_error = feature_error[feature_error["Frequency"] > 15]
    feature_error = feature_error[feature_error["Feature Weight"] > 0.1]
    #feature_error["h-v"] = feature_error["Horizontal Absolute Difference"] - feature_error["Vertical Absolute Difference"]
    #feature_error = feature_error[feature_error["h-v"]>0]

    return(feature_error)
```

```python
AFPT =  pd.read_csv('/Users/cmu-work/Code/11663AML/Midterm/actual_F_pred_T.csv')
AFPT_filtered = filter_feature(AFPT)
```

```python
AFPT_filtered.sort_values(by=["Horizontal Absolute Difference"], ascending=False)
```

```python
AFPT_filtered.sort_values(by=["Vertical Absolute Difference"], ascending=True)
```
