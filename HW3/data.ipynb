```python
def filter_feature(path):
    feature_error = pd.read_csv(path)
    feature_error = feature_error[feature_error["Frequency"] > 9]
    feature_error = feature_error[feature_error["Feature Weight"] > 0.05]
    feature_error["h-v"] = feature_error["Horizontal Absolute Difference"] - feature_error["Vertical Absolute Difference"]
    #feature_error = feature_error[feature_error["h-v"]>0]

    return(feature_error)
```

```python
# For NP
path = "/Users/cmu-work/lightside/HW3/export_NP(AP).csv"
NN_error = filter_feature(path)
NN_error.sort_values(by=["Horizontal Absolute Difference"], ascending=False)
```

```python
# for PN
path = "/Users/cmu-work/lightside/HW3/export_PN(AP).csv"
NN_error = filter_feature(path)
NN_error.sort_values(by=["Vertical Absolute Difference"], ascending=True)
```
